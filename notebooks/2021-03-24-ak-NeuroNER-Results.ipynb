# Goals of Care (NeuroNER Model Run)

**Goal:** Grab example sentences from [`2020-12-04-ak-NeuroNER-Results.ipynb`](2020-12-04-ak-NeuroNER-Results.ipynb).

```python
import pandas as pd
from cleaning.caregivers.neuroner import load_data
```

```python
df = load_data(keep_string_cols=True)
```

```python
df.groupby("HADM_ID")["CAR"].any().value_counts()
```

```python
df.groupby("HADM_ID")["LIM"].any().value_counts()
```

```python
df
```

```python
def get_extracted_text(result_string):
    results = result_string.splitlines()
    results = [r.split()[-1] for r in results]
    return " ".join(results)

get_extracted_text(df["RESULT_STRING_CAR"].iloc[0])
```

```python
df["RESULT_STRING_CAR"] = df["RESULT_STRING_CAR"].apply(get_extracted_text)
df["RESULT_STRING_LIM"] = df["RESULT_STRING_LIM"].apply(get_extracted_text)
```

```python
df["RESULT_STRING_CAR"].nunique()
```

```python
df["RESULT_STRING_LIM"].nunique()
```

```python
df
```

```python
samples_car = df[df["CAR"]]["RESULT_STRING_CAR"].sample(n=1000, random_state=0).reset_index(drop=True)
samples_lim = df[df["LIM"]]["RESULT_STRING_LIM"].sample(n=1000, random_state=0).reset_index(drop=True)
```

```python
samples_car.to_csv("2021-03-24-NeuroNER-Results-CAR-sample.csv", index=False)
samples_lim.to_csv("2021-03-24-NeuroNER-Results-LIM-sample.csv", index=False)
```
